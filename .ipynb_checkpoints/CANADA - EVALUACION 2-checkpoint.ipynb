```python
canada_n=canada.to_crs(5641)
# just Brazil

ciudad_canada=ciudad[ciudad.COUNTRY=='Canada']
ciudad_n=ciudad_canada.to_crs(canada_n.crs)

agua_n=agua.to_crs(canada_n.crs)
agua_n=gpd.clip(gdf=agua_n,mask=canada_n)
```

```python
# TODO EN CONJUNTO_N
base = canada_n.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
ciudad_n.plot(marker='+', color='red', markersize=15,ax=base)
agua_n.plot(edgecolor='blue', linewidth=0.5,ax=base)
```

```python
pais.to_file(os.path.join("maps","worldMaps.gpkg"), layer='countries', driver="GPKG")
ciudad.to_file(os.path.join("maps","worldMaps.gpkg"), layer='cities', driver="GPKG")
agua.to_file(os.path.join("maps","worldMaps.gpkg"), layer='rivers', driver="GPKG")
```

```python
# Specify the GeoPackage file path
gpkg_file = os.path.join("maps", "canada_n", "canadaMaps_5641.gpkg")

# Check if the directory exists and create it if it doesn't
os.makedirs(os.path.dirname(gpkg_file), exist_ok=True)

# Save each GeoDataFrame to the GeoPackage with a unique layer name
canada_n.to_file(gpkg_file, layer='country', driver="GPKG")
ciudad_n.to_file(gpkg_file, layer='cities', driver="GPKG")
agua_n.to_file(gpkg_file, layer='rivers', driver="GPKG")
```

```python
# Combine the GeoDataFrames
import pandas as pd

# Create an empty GeoDataFrame to store the combined data
all_data = gpd.GeoDataFrame()

# Append GeoDataFrames using pandas concat
all_data = pd.concat([all_data, canada_n, ciudad_n, agua_n], ignore_index=True)

# Plot the combined data
fig, ax = plt.subplots(figsize=(10, 10))
all_data.plot(ax=ax, facecolor="greenyellow", edgecolor='black', linewidth=0.4, markersize=15)
plt.show()
```

```python
# Plot each GeoDataFrame independently
# Save the GeoDataFrames to a single GeoPackage file

combined_gdf = gpd.GeoDataFrame(pd.concat([canada_n, ciudad_n, agua_n], ignore_index=True))

# Save the combined GeoDataFrame to a single GeoPackage file
combined_gdf.to_file("combined_data.gpkg", driver="GPKG")

# Plot each GeoDataFrame independently
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
canada_n.plot(ax=ax[0], facecolor="greenyellow", edgecolor='black', linewidth=0.4)
ciudad_n.plot(ax=ax[1], marker='+', color='red', markersize=15)
agua_n.plot(ax=ax[2], edgecolor='blue', linewidth=0.5)

plt.show()
```
